#### **Set the max and min values for x-axis and y-axis**

```python
fare = df.Fare[:100]

p_id = df.PassengerId[:100]

x_min, x_max, y_min, y_max = min(p_id), max(p_id), min(fare), max(fare)

print('Axis range: X = {0}    Y = {1}'.format((x_min, x_max),(y_min, y_max)))

plt.plot(p_id, fare)

plt.axis([x_min, x_max, y_min, y_max])

plt.show()
```

## **Subplots**

#### **Simple Subplot**

```python
x = np.linspace(-50, 50, 20)

plt.subplot(2, 2, 1) # num_row, num_col, subplot position
plt.plot(x, x, 'g--')

plt.subplot(2, 2, 2)
plt.plot(x, x**2, 'r:')

plt.subplot(223) 
plt.plot(x, x**3, 'b^')

plt.subplot(224)
plt.plot(x, x**4, 'y-.')

plt.show()
```

```python
x = np.linspace(-50, 50, 20)

plt.subplot(2, 2, 1) # num_row, num_col, subplot position
plt.plot(x, x, 'g--')

plt.subplot(2, 2, 2)
plt.plot(x, x**2, 'r:')

plt.subplot(2, 1, 2) 
plt.plot(x, x**3, 'b^')

plt.show()
```

#### **Grid Subplot**

```python
x = np.linspace(-50, 50, 20)

plt.subplot2grid((3,3), (0,0), colspan=2)
plt.plot(x, x**2)

plt.subplot2grid((3,3), (0,2), rowspan=2)
plt.plot(x, x**3)

plt.subplot2grid((3,3), (1, 0), rowspan=2)
plt.plot(x, x**4)

plt.subplot2grid((3,3), (1, 1))
plt.plot(x, x)

plt.subplot2grid((3,3), (2, 1), colspan=2)
plt.plot(x, x**5)
```

## **Multiple figures**

```python
x = np.linspace(-5, 5, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title('Square and Cube')
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title('y = x**4')
plt.subplot(122)
plt.plot(x, x**5)
plt.title('y = x**5')

plt.figure(1) # back to figure 1, curent subplot is 212 (bottom)
plt.subplot(212)
plt.plot(x, x, 'r--')

plt.show()
```

## **Drawing Text**

```python
x_min = -10
x_max = 10
data_point_n = 30

x = np.linspace(x_min, x_max, data_point_n)

px = x[data_point_n//2]
py = px**2

plt.plot(x, x**2, 'r--')

plt.plot(px, py, 'bo') # plotting the (px, py) point

plt.text(0, 85, 'Square function\ny=x^2', fontsize=10, color='blue', horizontalalignment='center')

plt.text(px-1 , py+1, "Beautiful point", ha="right", weight="heavy")

plt.text(px + 0.05, py - 0.3, "x=%0.2f\ny=%0.2f"%(px, py), rotation=-20, color='gray')

plt.show()
```

## **Just for Fun**

```python
# xkcd style

x_min = -10
x_max = 10
data_point_n = 30

x = np.linspace(x_min, x_max, data_point_n)

px = x[data_point_n//2]
py = px**2

with plt.xkcd():

  plt.plot(x, x**2, 'r--')

  plt.plot(px, py, 'bo') # plotting the (px, py) point

  plt.text(0, 85, 'Square function\ny=x^2', fontsize=10, color='blue', horizontalalignment='center')

  plt.text(px-1 , py+1, "Beautiful point", ha="right", weight="heavy")

  plt.text(px + 0.05, py - 0.3, "x=%0.2f\ny=%0.2f"%(px, py), rotation=-20, color='gray')

  plt.show()
```

## **Legend**

```python
x = np.linspace(-1.4, 1.4, 50)
plt.plot(x, x**2, "r--", label="Square function")
plt.plot(x, x**3, "g-", label="Cube function")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

## **Polar projection**

```python
radius = 1

two_pi_r = 2*np.pi*radius

theta = np.linspace(0, two_pi_r, 500)

plt.subplot(111, projection='polar')

plt.plot(theta, np.sin(5*theta), "r-.")

plt.plot(theta, 0.5*np.cos(20*theta), "g--")

plt.show()
```

## **Scatter plot**

```python
np.random.seed(42)  

x, y = np.random.rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
x, y, scale = np.random.rand(3, 100)
scale = 500 * scale ** 5
plt.scatter(x, y, s=scale)
plt.show()
```

```python
for color in ['red', 'green', 'blue']:
    n = 100
    x, y = np.random.rand(2, n)
    scale = 500.0 * np.random.rand(n) ** 5
    plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='red')

plt.grid(True)

plt.show()
```

## **Histograms**

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(212)
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.95)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
